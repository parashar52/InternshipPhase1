# 1) Write a python program to display all the header tags from wikipedia.org.

```python
#Send get request to the webpage server to get the source code of the page
url='https://en.wikipedia.org/wiki/Main_Page'
page = requests.get(url)
# page content
soup = BeautifulSoup(page.content)
# Scrap the Header tag
header_tags=[]

for header in soup.find_all(["h1","h2","h3","h4","h5","h6"]):
    header_tags.append(header.name+" "+header.text.strip())
    
#print all header_tags
header_tags

```

# 2) Write s python program to display list of respected former presidents of India(i.e. Name , Term of office) from https://presidentofindia.nic.in/former-presidents.htm

```python
# Send url request and download page content
presedent_soup=BeautifulSoup(requests.get("https://presidentofindia.nic.in/former-presidents.htm").content,"html.parser")
# Scrap President name
name=[i.text.split('\n')[1] for i in presedent_soup.find_all('div',class_="presidentListing")]
# Scrap terms of office
office_term=[i.text.split('\n')[2] for i in presedent_soup.find_all('div',class_="presidentListing")]
# Create DataFrame
former_President=pd.DataFrame({"Former Presidents":name,"Term of Office":office_term})
former_President
```

# 3) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:

a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

b) Top 10 ODI Batsmen along with the records of their team and rating.

c) Top 10 ODI bowlers along with the records of their team and rating.

# a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
#send get request to the webpage server to get the source code of the page
url = "https://www.icc-cricket.com/rankings/mens/team-rankings/odi"
page5 = requests.get(url)
# see content in page5
soup5 = BeautifulSoup(page5.content)
#scrape team names
team = soup5.find_all("span",class_='u-hide-phablet')
team_name = []
for i in team:
    team_name.append(i.text)
matches = [] #empty list
points = [] #empty list
ratings = [] #empty list
new_list = [] #empty list

for i in soup5.find_all("td",class_='rankings-block__banner--matches'): # first place team number of matches
    matches.append(i.text)
for i in soup5.find_all("td",class_='rankings-block__banner--points'):# first place team points
    points.append(i.text)
for i in soup5.find_all("td",class_='rankings-block__banner--rating u-text-right'):# first place team ratings
    ratings.append(i.text.replace("\n",""))
for i in soup5.find_all("td",class_='table-body__cell u-center-text'):# other teams number of matches and points
    new_list.append(i.text)
for i in range(0,len(new_list)-1,2):
    matches.append(new_list[i]) # other teams matches
    points.append(new_list[i+1]) # other teams points
for i in soup5.find_all("td",class_='table-body__cell u-text-right rating'):# other teams ratings
    ratings.append(i.text)
    
# Make data frame of top 10 ICC teams
icc=pd.DataFrame({})
icc['Team_name']=team_name[:10]
icc['Matches']=matches[:10]
icc['Points']=points[:10]
icc['Ratings']=ratings[:10]
icc    
```

# 5) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world :

i) Headline

ii) Time

iii) News Link

```python
# Send url request and download page source code
News_soup=BeautifulSoup(requests.get("https://www.cnbc.com/world/?region=world").content,"html.parser")
Headline=[i.text for i in News_soup.find_all(class_="LatestNews-headline") ] # Scrap Headlines
Time=[i.text for i in News_soup.find_all("span",class_="LatestNews-wrapper") ] # Scrap Time
# Scrap news Links
for links in News_soup.find_all("a",class_="LatestNews-headline"):
    links=links.get('href')
# Create data frame
news1=pd.DataFrame({'Headlines':Headline,'Time':Time,"Link":links})
news1
```

# 7) Write a python program to scrape mentioned details from dineout.co.in :

i) Restaurant name

ii) Cuisine

iii) Location

iv) Ratings

v) Image URL

```python
#send get request to the webpage server to get the source code of the page and download the page content
soup=BeautifulSoup(requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special').content,'html.parser')
name=[i.text.split(',')[0] for i in soup.find_all('div',class_='restnt-info cursor')] # Scrap Restaurant name
location=[i.text for i in soup.find_all('div',class_='restnt-loc ellipsis')] # Scrap location 
image=[i['data-src'] for i in soup.find_all("img", class_="no-img")] # scrap url of image
cuisine=[i.text.split('|')[1] for i in soup.find_all('span',class_="double-line-ellipsis")] # scrap cuisine
rating=[i.text for i in soup.find_all('div',class_='restnt-rating rating-4')] # scrap rating
df=pd.DataFrame({'Restaurant Name':name,'Cuisine':cuisine,'Location':location,"Rating":rating,"Image_url": image})
df
```
